```python
# Load the input Excel file with URLs
df_input = pd.read_excel('Input.xlsx')
df_input
```

```python
# Function to extract article text from a given URL
def extract_article_text(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract article title and text
        title = soup.find('title').get_text() if soup.find('title') else 'Untitled'
        article_text = ' '.join([p.get_text() for p in soup.find_all('p')])

        # Clean the text using regex (remove unwanted elements)
        article_text = re.sub(r'\[.*?\]', '', article_text)  # Remove text within square brackets

        return title, article_text
    else:
        return None, None
```

```python
# Create empty lists to store extracted data
titles = []
texts = []

# Iterate through rows in the input DataFrame
for index, row in df_input.iterrows():
    url = row['URL']

    # Extract article text
    title, article_text = extract_article_text(url)

    # Append the results to the lists
    titles.append(title)
    texts.append(article_text)

# Create a new DataFrame with extracted data
df_output = pd.DataFrame({
    'Title': titles,
    'Text': texts
})
```

```python
# Save the output to a new Excel file
df_output.to_excel('output_extraction.xlsx', index=False)
```

```python
import pandas as pd
import spacy
from textblob import TextBlob
import syllables
```

```python
# Load the spaCy English language model
nlp = spacy.load('en_core_web_sm')

# Load the extracted data Excel file
df_extracted_data = pd.read_excel('output_extraction.xlsx')
```

```python
# Function to compute linguistic features
def compute_linguistic_features(text):
    # Convert non-string input to empty string
    text = str(text) if text is not None else ''
    # Use spaCy for advanced linguistic analysis
    doc = nlp(text)

    # Compute positive, negative, polarity, and subjectivity scores using TextBlob
    blob = TextBlob(text)
    positive_score = blob.sentiment.polarity
    negative_score = -positive_score  # Negative score is the negative of the positive score
    polarity_score = blob.sentiment.polarity
    subjectivity_score = blob.sentiment.subjectivity

    # Compute average sentence length
    avg_sentence_length = len(list(doc.sents)) / len(list(doc.sents)) if len(list(doc.sents)) > 0 else 0

    # Compute percentage of complex words
    complex_words = [token for token in doc if token.is_alpha and token.is_stop]
    complex_word_count = len(complex_words)
    percentage_complex_words = (complex_word_count / len(doc)) * 100 if len(doc) > 0 else 0

    # Compute FOG index
    fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)

    # Compute average number of words per sentence
    avg_words_per_sentence = len(doc) / len(list(doc.sents)) if len(list(doc.sents)) > 0 else 0

    # Compute syllable per word
    syllables_per_word = syllables.estimate(text) / len(doc) if len(doc) > 0 else 0

    # Compute personal pronoun count
    personal_pronoun_count = len([token for token in doc if token.pos_ == 'PRP' or token.pos_ == 'PRP$'])

    # Compute average word length
    avg_word_length = sum(len(token.text) for token in doc if token.is_alpha) / len([token for token in doc if token.is_alpha])

    return positive_score, negative_score, polarity_score, subjectivity_score, avg_sentence_length, \
           percentage_complex_words, fog_index, avg_words_per_sentence, complex_word_count, \
           len(doc), syllables_per_word, personal_pronoun_count, avg_word_length

# Apply function to each row in the DataFrame
df_extracted_data[['Positive_Score', 'Negative_Score', 'Polarity_Score', 'Subjectivity_Score',
                   'Avg_Sentence_Length', 'Percentage_Complex_Words', 'FOG_Index', 'Avg_Words_Per_Sentence',
                   'Complex_Word_Count', 'Word_Count', 'Syllables_Per_Word', 'Personal_Pronoun_Count',
                   'Avg_Word_Length']] = df_extracted_data['Text'].apply(compute_linguistic_features).apply(pd.Series)
```

```python
# Save the analyzed data to a new Excel file
df_extracted_data.to_excel('output_analysis_extended.xlsx', index=False)
```
